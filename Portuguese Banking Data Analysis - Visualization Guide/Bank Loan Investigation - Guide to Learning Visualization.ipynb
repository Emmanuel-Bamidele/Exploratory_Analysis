## 01 Line Graph

```python
#Example 1 - Simples 1 line graph 
#Investigate how loan balance vary with age

ax = sns.lineplot(x='age', y = 'balance', data = df).set(title='Investigate Loan Balance by Age', xlabel = 'Age', ylabel='Loan Balance')
#To customize axes labels to be different from column names

#Notes: error bands show the confidence intervals
```

```python
#Example 2 - Adding Categories
#Categorize by education

ax = sns.lineplot(x='age', y = 'balance', hue = 'education', data = df).set(title='Investigate Loan Balance by Age', xlabel = 'Age', ylabel='Loan Balance')
```

```python
#Example 3 - Categorize and add style to lines
#category used is education

ax = sns.lineplot(x='age', y = 'balance', hue = 'education', style = 'education', data = df).set(title='Investigate Loan Balance by Age', xlabel = 'Age', ylabel='Loan Balance')
```

```python
#Example 4 - Increase size of example 3

#increase the size
sns.set(rc={'figure.figsize':(10,8)})

ax = sns.lineplot(x='age', y = 'balance', hue = 'education', style = 'education', data = df).set(title='Investigate Loan Balance by Age', xlabel = 'Age', ylabel='Loan Balance')
```

```python
#Example 5 - remove error bars
#ci is confidence intervals

ax = sns.lineplot(x='age', y = 'balance', hue = 'education', style = 'education', data = df, ci=None).set(title='Investigate Loan Balance by Age', xlabel = 'Age', ylabel='Loan Balance')
```

```python
#Example 6 - add markers

ax = sns.lineplot(x='age', y = 'balance', hue = 'education', style = 'education', data = df, ci=None, markers=True).set(title='Investigate Loan Balance by Age', xlabel = 'Age', ylabel='Loan Balance')
```

```python
#Example 7 - Categorize with Marital Status

ax = sns.lineplot(x='age', y = 'balance', hue = 'marital_status', style = 'marital_status', data = df, ci=None).set(title='Investigate Loan Balance by Age', xlabel = 'Age', ylabel='Loan Balance')
```

## 02 Bar Plots

```python
#Example 1 - Total loan by jobs

ax = sns.barplot(x="job", y="balance", data=df)

#Notes:
#1 - there will lines indicating the confidence internal
#2 - It will calculate the mean by default

df[['job', 'balance']].groupby('job', as_index = False).agg({'balance':'mean'})
ax.set_xticklabels(ax.get_xticklabels(), rotation=45);
```

```python
#Example 2 - Removing Confidence Intervals

ax = sns.barplot(x="job", y="balance", data=df, ci=None)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45);
```

```python
#Example 3 - Categorize by education

ax = sns.barplot(x="job", y="balance", data=df, ci=None, hue = "education")
ax.set_xticklabels(ax.get_xticklabels(), rotation=45);
```

## 03. Histogram

```python
#Example 1 - Investigate distribution of age

x = df['age'].values

sns.distplot(x, color='blue');

#The curve is imbalance, skewed to the mean which is at the left
```

```python
# Example 2 - Investigating age distribution, adding Mean

x = df['age'].values

sns.distplot(x, color='blue');

#Calculating the mean
mean = df['age'].mean()

#plotting the mean
plt.axvline(mean, 0,1, color='red')
```

```python
# Example 3 - Investigating days distribution, adding Mean

x = df['campaign'].values

sns.distplot(x, color='blue');
ax.set_xlim(0,20)

#Calculating the mean
mean = df['campaign'].mean()

#plotting the mean
plt.axvline(mean, 0,1, color='red')
```
