```python
import os
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Import dataset
data_file = "suicide.csv"
df_world = pd.read_csv(data_file)
df_world.head()
```

```python
#Sort dataframe to obtain only dataset with Southafrica
df_SA = df_world.loc[df_world['country'] == 'South Africa']
df_SA.head()
```

```python
#Restart index after sorting the dataframe
df_SA = df_SA.reset_index(drop=True)
df_SA.head()
```

```python
#Check the columns of the dataframe
df_SA.columns
```

```python
#Rename the columns for easy use
#Note: GDP is in dollars
df_world.columns = ['country', 'year', 'sex', 'age', 'suicides_number', 'population', 'suicides_per_100k', 'country_year', 'HDI_for_year', 'GDP_for_Year', 'GDP_Per_Capital', 'generation']
df_SA.columns = ['country', 'year', 'sex', 'age', 'suicides_number', 'population', 'suicides_per_100k', 'country_year', 'HDI_for_year', 'GDP_for_Year', 'GDP_Per_Capital', 'generation']
df_SA.head()
```

```python
#save South Africa (SA) data
df_SA.to_csv('South_Africa_Suicide_Data.csv')
```

### Compare South Africa and the rest of the world

```python
#compare suicide rate in South Africa and the rest of the world

sns.set(rc={'figure.figsize':(8,6)})

ax1 = sns.lineplot(x = df_SA['year'], y = df_SA['suicides_number'], color = 'blue', label = 'South Africa').set(title='Comparison of South Africa Suicide with the rest of the World', xlabel = 'Year', ylabel = 'Suicide Number')
ax2 = sns.lineplot(x = df_world['year'], y = df_world['suicides_number'], color = 'red', label = 'World').set(title='Comparison of South Africa Suicide with the rest of the World', xlabel = 'Year', ylabel = 'Suicide Number')

plt.xlim(1985, 2016);
plt.legend();
```

### Investigate South Africa Annual Suicide Number

```python
#Investigation of suicide rate in South Africa and categorizing by age

ax1 = sns.lineplot(x = df_SA['year'], y = df_SA['suicides_number'], hue = df_SA['age'], ci=False).set(title='Investigation of suicide rate in South Africa and categorizing by age', xlabel = 'Year', ylabel = 'Suicide Number')

plt.xlim(1985, 2016);
```

```python
#Investigation of suicide rate in South Africa and categorizing by age

ax1 = sns.barplot(x = df_SA['age'], y = df_SA['suicides_number'], ci=False).set(title='Investigation of suicide rate in South Africa by age', xlabel = 'Age', ylabel = 'Suicide Number')
```

```python
#Investigation of suicide rate in South Africa and categorizing by sex

ax1 = sns.lineplot(x = df_SA['year'], y = df_SA['suicides_number'], hue = df_SA['sex'], ci=False).set(title='Investigation of suicide rate in South Africa and categorizing by sex', xlabel = 'Year', ylabel = 'Suicide Number')

plt.xlim(1985, 2016);
```

```python
#Investigation of suicide rate in South Africa and categorizing by sex

# Separating and adding the number of suicides by sex
df_SA_sex_percent = df_SA.groupby('sex')['suicides_number'].sum()

df_SA_sex_percent

pie_color = ['#83cceb', '#a7ebe6']
sex_labels = ['Female', 'Male']

plt.pie(df_SA_sex_percent, 
        labels=sex_labels,
        autopct='%.1f%%',
        shadow=True,
        colors=pie_color,
        explode=[0.1, 0]);
```

### Investigate Suicide Number as a function of GDP

```python
#Investigation of suicide rate in south Africa as a function of GDP per capital

ax1 = sns.lineplot(x = df_SA['GDP_Per_Capital'], y = df_SA['suicides_number']).set(title='Investigation of suicide rate in south Africa as a function of GDP per capital', xlabel = 'GDP Per Capital', ylabel = 'Suicide Number')
```

```python
#Investigation of suicide rate in south Africa as a function of GDP per year

ax1 = sns.lineplot(x = 'GDP_for_Year', y = 'suicides_number', data = df_SA).set(title='Investigation of suicide rate in south Africa as a function of GDP per year', xlabel = 'GDP Per Year', ylabel = 'Suicide Number')

plt.xticks(rotation=45, ha='right', fontweight='light', fontsize='small');
```

```python
#Investigation of suicide rate in south Africa as a function of GDP per year - bar plot

ax1 = sns.barplot(x = 'GDP_for_Year', y = 'suicides_number', data = df_SA, ci=None).set(title='Investigation of suicide rate in south Africa as a function of GDP per year', xlabel = 'GDP Per Year', ylabel = 'Suicide Number')

plt.xticks(rotation=45, ha='right', fontweight='light', fontsize='small');
```

### Investigating South Africa Suicide Rate by Generation

- Boomers (1946 - 1964) 
- Generation X (1961 - 1981) 
- Millennial Generation (1981 - 1996)

```python
# Grouping the number of suicides by generation
colors = ['y', 'm', 'b', 'c', 'g', 'r']
generation = df_SA.groupby('generation')['suicides_number'].sum().sort_values()

# Arranging the number of suicides by generation
gen_SA = generation.index


# Plotting a graph showing the total number of suicides per generation
plt.bar(gen_SA, generation.values,
       color=colors);

plt.xticks(rotation=45, ha='right', fontweight='light', fontsize='large');
```
